# LA LIGA dataset 2005 - 2020


```python
#import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#reading the dataset 
laliga_data = pd.read_csv(r"C:\Users\trevo\laliga21-22.csv")
laliga_data.head(19)
```

```python
laliga_data.describe()
```

```python
# win percentage of top 4
win_perc = laliga_data['W'].astype(int) / 38 * 100
top_4 = laliga_data.nlargest(4, 'Pts')

#Calculate the win percentage
top_4['Win Percentage'] = win_perc

#Create a bar plot
sns.barplot(x='Squad', y='Win Percentage', data=top_4)

# Set the title and axes labels
plt.title('Win Percentage of Top 4 Teams')
plt.xlabel('Squad')
plt.ylabel('Win Percentage')

# Display the plot
plt.show()
```

```python
#A new dataframe containing GF and GA for each team
gf_ga = top_4[['Squad', 'GF', 'GA']]

#Create a scatter plot
sns.scatterplot(x='GF', y='GA', hue='Squad', data=gf_ga)

# Set the title and axes labels
plt.title('GF vs GA for Top 4 Teams')
plt.xlabel('GF')
plt.ylabel('GA')

# Display the plot
plt.show()
```

```python
#A new dataframe containing GA and shots per game for each team
ga_spg = top_4[['Squad', 'GF', 'shots per']]

#Create a scatter plot
sns.scatterplot(x='shots per', y='GF', hue='Squad', data=ga_spg)

# Set the title and axes labels
plt.title('GF vs Shots per Game for Top 4 Teams')
plt.xlabel('Shots per Game')
plt.ylabel('GF')

# Display the plot
plt.show()
```

```python
# a new dataframe containing points per match
Pts_per_match = laliga_data['Pts'].astype(int) / laliga_data['MP'].astype(int)
top_4 = laliga_data.nlargest(4, 'Pts')

top_4['Points per Match Played'] = Pts_per_match

#Create a bar plot
sns.barplot(x='Squad', y='Points per Match Played', data=top_4)

# Set the title and axes labels
plt.title('Points per game of Top 4 Teams')
plt.xlabel('Squad')
plt.ylabel('Points per game')

# Display the plot
plt.show()
```
